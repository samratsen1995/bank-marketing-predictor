```python
import pandas as pd
df=pd.read_csv("Bank_Marketing_Original_Data.csv")
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
df.shape
```

```python
transformer=ColumnTransformer(transformers=[("tr1",OrdinalEncoder(categories=[["unknown","primary","secondary","tertiary"]]),["Education"]),("tr2",OneHotEncoder(),["Job","Marital","Default","Housing","Loan","contact","Month","poutcome"]),("tr3",SimpleImputer(strategy="constant",fill_value=5),["Age","Balance"])],remainder="passthrough")
```

```python
transformer
```

```python
x=df.iloc[:,:-1]
```

```python
y=df["TARGET"]
```

```python
x
```

```python
y
```

```python
x=transformer.fit_transform(x)
```
